#Importing the Necessary Libraries

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
```

# Displaying the Original Image

```python
img = cv2.imread("/content/received_380633059401712.jpeg")

plt.imshow(img)
plt.title("Original Image")
plt.show()
```

#Converting the Original Image to Grayscale and Displaying It

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Gray Image")
plt.show()
```

#Applying Gaussian Blur to Grayscale Image for a Softened Effect

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)

plt.imshow(blur, cmap='gray')
plt.title("Soft Blur Image")
plt.show()
```

#Detecting Edges in the Blurred Image using Canny Edge Detection

```python
edges = cv2.Canny(blur, 60, 100)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detected Image")
plt.show()
```

# Creating a Pencil Sketch Effect by Inverting Edges in the Image

```python
edges = cv2.bitwise_not(edges)

cv2.imwrite("sketch.jpg", edges)

weight = cv2.countNonZero(edges)
print("Sketch image saved with weight:", weight)

plt.imshow(edges, cmap='gray')
plt.title("Pencil Sketch Image")
plt.axis('off')
plt.show()
```
