```python
import matplotlib.pyplot as plt
from PIL import Image
from stegano import lsb
```

# Displaying the Original Image

```python
image_path = '/content/person.jpg'
original_image = Image.open(image_path)

plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

# Image Steganography: Encoding a Secret Message in an Image using Least Significant Bit (LSB)

```python
def encode_message(original_image_path, secret_message, encoded_image_path):
    original_image = Image.open(original_image_path)
    secret_image = lsb.hide(original_image, secret_message)
    secret_image.save(encoded_image_path)

    print("Message successfully encoded.")


secret_message = "Hello, this is Zahid! Now tell me who you are?"
encoded_image_path = '/content/encoded_image.png'

encode_message(image_path, secret_message, encoded_image_path)
```

# Image Steganography: Decoding a Secret Message from an Encoded Image using Least Significant Bit (LSB)

```python
def decode_message(encoded_image_path):
    encoded_image = Image.open(encoded_image_path)
    decoded_message = lsb.reveal(encoded_image)

    return decoded_message

decoded_message = decode_message(encoded_image_path)
print("Decoded Message:", decoded_message)
```

# Image Steganography: Visualizing Original, Encoded, and Decoded Images with Message

```python
plt.figure(figsize=(15,6))

plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

encoded_image = Image.open(encoded_image_path)
plt.subplot(1, 3, 2)
plt.imshow(encoded_image)
plt.title('Encoded Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.text(0.1, 0.5, f'Decoded Message:\n{decoded_message}', fontsize=10, ha='left', va='center')
plt.axis('off')

plt.show()
```
