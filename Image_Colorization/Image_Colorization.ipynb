```python
import cv2
import matplotlib.pyplot as plt
import fastai
from deoldify import device
from deoldify.device_id import DeviceId
from deoldify.visualize import *
from pathlib import Path
```

# Visualizing the Original Image

```python
image_path = '/content/person.jpg'
original_image = cv2.imread(image_path)

plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

# Converting Image to Grayscale

```python
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('gray_image.jpg', gray_image)
```

## Setting Torch Device Based on GPU Availability

```python
torch_device = torch.device(DeviceId.CPU)
if torch.cuda.is_available():
    torch_device = torch.device(DeviceId.GPU_0)
```

# Initializing DeOldify Image Colorizer with Artistic Mode

```python
colorizer = get_image_colorizer(artistic=True)
```

# Applying DeOldify Colorization to Grayscale Image

```python
input_path = Path("/content/gray_image.jpg")

colorizer.plot_transformed_image(path=input_path,
                                 render_factor=40,
                                 display_render_factor=False,
                                 figsize=(8,8),
                                 results_dir=Path('/content/result_images'))
```
